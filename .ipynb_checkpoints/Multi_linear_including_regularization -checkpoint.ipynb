# Regularization 

When we use the Linear Regression model there are chances that the model will overfit the given training data set. Regularization helps in reducing overfitting by penalizing coefficients.

First, we will try to understand the causes of overfitting in the linear regression model.

The model is too complex

When the model is going to have too many parameters and model parameters are having high values then it is said to be a complex model.

If the model is complex, it will have high Variance (Sensitivity ), resulting in overfitting.

let’s take an example of simple linear regression

Red points are training data and Yellow points are testing data. The line fits perfectly to training data but not too testing data. The model is not generalised. Linear Regression model is having m (slope)value to be high as the slope of the line is high so this result in a high variance model.

In y=mx+c if the m value is higher than the small change in x there is a big change in y. if the slope of the line would have less in that case there would be less overfitting.

```python
Image(url="https://miro.medium.com/max/700/1*pQLaBxZ32grRSgjxtSM-LA.png")
```

```python
def adj_r2(x,y,model):
    r2 = model.score(x,y)
    n = x.shape[0]
    p = x.shape[1]
    adjusted_r2 = 1-(1-r2)*(n-1)/(n-p-1)
    return adjusted_r2
```

#Linear Regression 

```python
regression.score(x_train,y_train) # R2 for training dataset 
```

```python
adj_r2(x_train,y_train) #adjusted R2 for training dataset 
```

2 score is 84.15% and adj r2 is 83.85% for our training dataset 

```python
regression.score(x_test,y_test) #Test Accurtacy (r-Square value)
```

```python
adj_r2(x_test,y_test,regression) #Test Accurtacy (Adjusted r-Square value)
```

 R-Square score is 75.34% and adj R-Square is 73.87% for our Testing dataset

```python
mean_squared_error(y_test,regression.predict(x_test))
```

# Lasso Regression 

# L1: LASSO (Least Absolute Shrinkage Selector Operator)

in LASSO regularization we make a change in cost function by adding penalizing term.

Lasso Cost Function =Sum of Error + Sum of the absolute value of coefficients.

LASSO Cost function will find a predicted line which is having less m value and also less error. It can also eliminate some features by making the m value zero.

```python

Image(url= "https://miro.medium.com/max/556/1*-keaVhZSasb_rvg2OgxRkQ.png")
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso=Lasso()
parameters={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100,200,300,1.1,0.9,0.8,0.7,0.5,0.4,0.3,0.2,0.33]}
lasso_regressor=GridSearchCV(lasso,parameters,scoring='neg_mean_squared_error',cv=5)

lasso_regressor.fit(X,y)
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

Best alpha value is 0.00000001 & MSE is 0.00400

```python
lasso_alpha=0.00000001
```

```python
lasso_reg = Lasso(lasso_alpha)
lasso_reg.fit(x_train, y_train)
```

```python
r2=lasso_reg.score(x_train,y_train)
print(r2)
```

```python
adj_r2(x_train,y_train,lasso_reg)
```

```python
adj_r2(x_test,y_test,lasso_reg)
```

# Ridge Regression

# L2: Ridge

Ridge cost function: Sum of Error + Sum of the squares of coefficients

Ridge shrink the parameter it helps in preventing multi colinearity.

```python
Image(url= "https://miro.medium.com/max/549/1*GwNOQbkMHuAxTElptY3lZg.png")
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

ridge=Ridge()
parameters={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100,101,220,219,215,210,207,208,206,205]}
ridge_regressor=GridSearchCV(ridge,parameters,scoring='neg_mean_squared_error',cv=5)
ridge_regressor.fit(X,y)
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

```python
Ridge_alpha=0.000000000000001 #best alpha value
```

```python
ridge_reg = Lasso(Ridge_alpha)
ridge_reg.fit(x_train, y_train)
```

```python
r2=ridge_reg.score(x_train,y_train)
print(r2)
```

```python
adj_r2(x_train,y_train,ridge_reg)
```

```python
adj_r2(x_test,y_test,ridge_reg)
```

# Elastic net

 Elastic net uses the combination of LASSO & Ridge.
 
Elastic net: Sum of errors + Sum of the absolute value of coefficients + Sum of the squares of coefficients.






```python
Image(url= "https://miro.medium.com/max/495/1*MXHzoLqURcTFy7ITIZqiQg.png")
```

```python
from sklearn.linear_model  import  ElasticNetCV
elasticCV = ElasticNetCV(alphas = None, cv =5)

elasticCV.fit(x_train, y_train)
```

```python
elastic_alpha=elasticCV.alpha_
elastic_alpha
```

```python
from sklearn.linear_model  import  ElasticNet
elasticnet_reg = ElasticNet(alpha = elastic_alpha)
elasticnet_reg.fit(x_train, y_train)
```

```python
elasticnet_reg.score(x_train, y_train)
```

```python
elasticnet_reg.score(x_test, y_test)
```

```python
adj_r2(x_train,y_train,elasticnet_reg)
```

```python
adj_r2(x_test, y_test,elasticnet_reg)
```


 we can see by using different type of regularization, we still are getting the same MSE & r2 score. That means our  model has been well trained over the training data
